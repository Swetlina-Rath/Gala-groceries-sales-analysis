```python
sales_by_category = df.groupby('category')['total'].sum()
plt.bar(sales_by_category.index, sales_by_category.values)
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Product Category')
plt.xticks(rotation=90)
plt.show()
```

most of the cells are from kitchen category.

```python
df['timestamp'] = df['timestamp'].astype('datetime64[ns]')
```

```python
df['timestamp'].dtype
```

```python
df['date_only'] = df['timestamp'].dt.date.astype('datetime64[ns]')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['date_only'])
sales_over_time = df.groupby('date_only')['total'].sum()
plt.plot(sales_over_time.index, sales_over_time.values)
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.title('Sales Trend over Time')
plt.xticks(rotation=90)
plt.show()
```

maximum sales have taken place on 3rd february, 2022.

```python
# bar chart of category
ax = sns.countplot(x = 'category',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(38,5)})
```

fruits are sold in most quantity.

```python
sales_cat = df.groupby(['category'], as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False)

sns.barplot(x = 'category',y= 'quantity' ,data = sales_cat)
```

```python
From the above graphs we can see that most of the buyers buy fruits and even the quantity of fruits bought is highest among all.
```

```python
sales_cus = df.groupby(['customer_type'], as_index=False)['quantity'].sum().sort_values(by='customer_type', ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'customer_type',y= 'quantity' ,data = sales_cus )
```

```python
most quantity of products has been brought by standard customer type.
```

```python
sales_pri = df.groupby(['category'], as_index=False)['unit_price'].sum().sort_values(by='unit_price', ascending=False)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x = 'category',y= 'unit_price' ,data = sales_pri )
```

```python
the unit price of kitchen is more than that of all categories.
```

```python
sales_tot = df.groupby(['customer_type'], as_index=False)['total'].sum().sort_values(by='total', ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x = 'customer_type',y= 'total' ,data = sales_tot )
```

```python
from the above graph it can be concluded that non members spend most money in buying the products.
```

```python
sales_pay = df.groupby(['payment_type'], as_index=False)['total'].sum().sort_values(by='total', ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x = 'payment_type',y= 'total' ,data = sales_pay )
```

```python
it can be clear that most of the payments are done using cash
```

```python
df['customer_type'] = df['customer_type'].astype(str)
```

```python
plt.figure(figsize=(38, 10))
sns.countplot(x='category', hue='customer_type',data=df)

# Add title and labels
plt.title('Number of Sales by Customer Type and Category')
plt.xlabel('category')
plt.ylabel('count')
plt.legend(title='customer type')

# Show the plot
plt.show()
```

from the above observation its clear that most of the standard customer type spend on buying fruits.
